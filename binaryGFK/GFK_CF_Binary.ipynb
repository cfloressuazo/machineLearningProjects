highest percentage for missing values is for v16 with **~58%**. There rest of variables have missing values that we can replace based on their data distribution.

A simple correlation heatmap is shown above. This map describes the degree of nullity relationship between the different features. The range of this nullity correlation is from -1 to 1 (-1 ≤ R ≤ 1). Features with no missing value are excluded in the heatmap. If the nullity correlation is very close to zero (-0.05 < R < 0.05), no value will be displayed. Also, a perfect positive nullity correlation (R=1) indicates when the first feature and the second feature both have corresponding missing values while a perfect negative nullity correlation (R=-1) means that one of the features is missing and the second is not missing.

From both visualizations these are my findings:
- v13 with v15 are fully correlated in terms of nullity correlation, this mean the messing value for either variable affects in the same manner the other.
- v4 is highly correlated with v13 and v15, same as above, which makes me think they explain same phenomen.
- other variables can be then be imputed by either the mean or median with no major impact.
- is the variable that contains more missing values.

I will first focus on replacing these variable's missing points, starting with the numerical variables and then categorical, and then moving to v16 as it seems to have a major issue or impact on filling out these points.

```python
# v2
# The most common and simplistic approach is to fill it with either the mean or the median of this variable.
print(df_train.v2.describe())
df_train_check_distribution=df_train.drop(df_train[pd.isnull(df_train.v2)].index)
fig, axes = plt.subplots(1, 2, figsize=(17,3.5))
axes[0].boxplot(df_train_check_distribution.v2, showmeans=True,vert=False,)
axes[0].set_xlim(0,max(df_train_check_distribution.v2))
axes[0].set_title('Boxplot For v2')
axes[1].hist(df_train_check_distribution.v2, cumulative=False, bins=15) # tried with 10, 15, 20 and 30
axes[1].set_title("v2 histogram")
axes[1].set_xlim((min(df_train_check_distribution.v2), max(df_train_check_distribution.v2)))
{"Mean":np.nanmean(df_train.v2),"Median":np.nanmedian(df_train.v2),"Standard Dev":np.nanstd(df_train.v2)}
```

```python
# we can see a right skewed distribution on the v2 and small difference between the
# mean and median. there is a certain level of dispercity and explained through the standard dev (nearly 2x mean)
# so in this case i will use the median to replace nan
df_train['v2'].fillna(df_train['v2'].median(), inplace=True)
print(df_train['v2'].median())
```

```python
# v15 and v13 are fully correlated on nulity so I will need to use a method consistent for both,
# at the same time these are highly correlated with the categorical variable v4, so it would be 
# good for the analysis to check on the distributon of these 3 variables and how the interact with each other.
print(df_train[['v4', 'v13', 'v15']].groupby(df_train['v4']).describe())
#
# same amount of missing values for both variables (v13 and v15) on each category. I have no clue on why
# the values on v4 are missing, and their distribution might bias the further learning process if i decide
# to replace with the mode.
print("\nmissing values for v4 where v15/v13 is null and v4 is not null: {}\n".format(df_train['v4'].loc[df_train['v15'].isnull() & df_train['v4'].notnull()].count()))
print(df_train[df_train['v4'].isnull()].groupby(df_train['classLabel']).count())
#
# all missing values on v4 are of class "yes".
df_train['v4'].groupby(df_train['classLabel']).count()
#
# i will delete the missing rows as filling them out in regarding to the "yes" class might produced
# more imbalance dataset.
df_train.dropna(subset=['v4'], inplace=True)
#
# let's look at the missing values per column, and their percentages also
print("-------- MISSING DATA PERCENTAGES/SUM --------")
print(df_train.isnull().sum()/df_train.classLabel.size * 100)
print(df_train.isnull().sum())
#
# let's check the category (v4) and their distribution on each variable (v13, v15)
res = df_train[df_train['v13'].isnull()][['v13', 'v4']].fillna('1')
print(res.groupby('v4').count())
res = df_train[df_train['v15'].isnull()][['v15', 'v4']].fillna('1')
print(res.groupby('v4').count())
#
# 33 are for "u" and 3 are for "y". I will replace for the median of v13 and v15 per category
df_train['v13'].loc[df_train['v4'] == 'y'] = df_train['v13'].loc[df_train['v4'] == 'y'].fillna(df_train[['v4', 'v13', 'v15']].groupby(df_train['v4']).median()['v13']['y'])
df_train['v13'].loc[df_train['v4'] == 'u'] = df_train['v13'].loc[df_train['v4'] == 'u'].fillna(df_train[['v4', 'v13', 'v15']].groupby(df_train['v4']).median()['v13']['u'])
df_train['v15'].loc[df_train['v4'] == 'y'] = df_train['v15'].loc[df_train['v4'] == 'y'].fillna(df_train[['v4', 'v13', 'v15']].groupby(df_train['v4']).median()['v15']['y'])
df_train['v15'].loc[df_train['v4'] == 'u'] = df_train['v15'].loc[df_train['v4'] == 'u'].fillna(df_train[['v4', 'v13', 'v15']].groupby(df_train['v4']).median()['v15']['u'])
print(df_train[['v4', 'v13', 'v15']].groupby(df_train['v4']).median()['v13']['y'])
print(df_train[['v4', 'v13', 'v15']].groupby(df_train['v4']).median()['v13']['u'])
print(df_train[['v4', 'v13', 'v15']].groupby(df_train['v4']).median()['v15']['y'])
print(df_train[['v4', 'v13', 'v15']].groupby(df_train['v4']).median()['v15']['u'])
#
# let's look at the missing values per column, and their percentages also
print("-------- MISSING DATA PERCENTAGES/SUM --------")
print(df_train.isnull().sum()/df_train.classLabel.size * 100)
print(df_train.isnull().sum())
#
# plotting the histogram for both variables
fig, axes = plt.subplots(1, 2, figsize=(17,3.5))
axes[0].hist(df_train.v13)
axes[0].set_title('Histogram For v13')
axes[1].hist(df_train.v15)
axes[1].set_title("Histogram For v15")
```

```python
#
# histogram for variables are not screewed with the median as the replacement, as these where smaller values also,
# replacing v13 and v15 with the meadian wont add any bias and would help to preserve the original distribution
#
# to replace v1 I would need to analyzie the impact of these variable with the target variable. as this is a
# categorical variable I can work on the simple statistic and then move towards a feature engineering approach.
print("missing percentage for v1: {0}".format((1-df_train['v1'].describe()[0]/df_train['v1'].size)*100))
print(df_train['classLabel'].groupby(df_train['v1']).describe())
#
# 1247 - 1168 = 79 are "no" for v1 a
# 2350 - 2160 = 190 are "no" for v1 b
print("-------- V1 NULL GROUP BY CLASS LABEL --------")
print(df_train['v1'].isnull().groupby(df_train['classLabel']).sum())
#
# as most null values for v1 are for "yes" class label, i will replace with the frequent value (b)
df_train['v1'].fillna(df_train['v1'].value_counts().index[0], inplace=True)
#
# transform classlabel into integer representation (binary 1 or 0)
class_label_bin = {"classLabel": {"no.": 0, "yes.": 1}}
df_train.replace(class_label_bin, inplace=True)
#
# double check missing values
print("-------- MISSING DATA PERCENTAGES/SUM --------")
print(df_train.isnull().sum()/df_train.classLabel.size * 100)
```

now i have to handle the 57% of missing values for v16. 

1. one option is to drop the entire column and not use it in the learning process, but this might let to loss of information or data. 

2. second option is to replace with a new category "U" as unknown but this adds another feature to the model while encoding, which may result in poor performance.

3. third option is to use the values of other columns to impute the missing values in v16. imputing the missing variable is an improvement as long as the bias from the same is smaller than the  omitted variable bias.

```python
pd.options.mode.chained_assignment = None
from sklearn.ensemble import RandomForestClassifier

pred_impute = df_train[df_train['v16'].isnull()]
train_impute = df_train.dropna()
train_impute['v1'] = train_impute['v1'].astype('category').cat.codes
train_impute['v4'] = train_impute['v4'].astype('category').cat.codes
train_impute['v8'] = train_impute['v8'].astype('category').cat.codes
train_impute['v9'] = train_impute['v9'].astype('category').cat.codes
train_impute['v11'] = train_impute['v11'].astype('category').cat.codes
train_impute['v12'] = train_impute['v12'].astype('category').cat.codes
#
# data preparation for training
x_pred = train_impute.drop('v16', axis=1)
y_pred = train_impute['v16']
#
# simple random forect classifier for v16 imputation
clf = RandomForestClassifier(max_depth=128, random_state=4, verbose=True)
# fit model
clf.fit(x_pred, y_pred)
#
# impute missing values
pred_impute = pred_impute.drop('v16', axis=1)
pred_impute['v1'] = pred_impute['v1'].astype('category').cat.codes
pred_impute['v4'] = pred_impute['v4'].astype('category').cat.codes
pred_impute['v8'] = pred_impute['v8'].astype('category').cat.codes
pred_impute['v9'] = pred_impute['v9'].astype('category').cat.codes
pred_impute['v11'] = pred_impute['v11'].astype('category').cat.codes
pred_impute['v12'] = pred_impute['v12'].astype('category').cat.codes
#
# prediction with missing values
pred_v16 = clf.predict(pred_impute).tolist()
#
# impute values
pred_impute['v16'] = pred_v16
# back to the original dataframe
df_train['v16'].loc[pred_impute.index] = pred_impute.v16
#
# double check missing values
print("-------- MISSING DATA PERCENTAGES/SUM --------")
print(df_train.isnull().sum()/df_train.classLabel.size * 100)
```

# Feature Engineering
now that i have my dataset ready with no null values, i will proceed to train the binary classifier.

1. first transform all categorical variables into right format, for this i will use one hot encoded with get_dummies
2. get correlation with the target variable
3. select best features
4. train a random forect classifer

```python
df_cat = pd.get_dummies(df_train[['v1', 'v4', 'v8', 'v9', 'v11', 'v12', 'v16']], prefix=['v1', 'v4', 'v8', 'v9', 'v11', 'v12', 'v16']).astype(int)
df_train_cat = pd.concat([df_train, df_cat], axis=1, join_axes=[df_train.index])
df_train_cat = df_train_cat.drop(columns=['v1', 'v4', 'v8', 'v9', 'v11', 'v12', 'v16'], axis=1)
print(df_train_cat.dtypes)
```

```python
# printing correlation as to define features that might add noise instead supporting the learning process
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
from scipy import stats
import itertools

df_correlation=df_train_cat
upper_triangle = np.zeros_like(df_correlation.corr(), dtype = np.bool)
upper_triangle[np.triu_indices_from(upper_triangle)] = True
f, ax = plt.subplots(figsize = (20, 10))
sns.heatmap(df_correlation.corr(),ax=ax,mask=upper_triangle,annot=True, fmt='.2f',linewidths=0.5,cmap=sns.diverging_palette(10, 133, as_cmap=True))
```

```python
df_train_cat.shape
```

```python
# Choosing best features based on RFE
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestRegressor

X = df_train_cat.drop(columns=['classLabel'], axis=1)
y = df_train['classLabel']

# feature extraction
model = RandomForestRegressor(n_estimators=128, 
                             criterion='mse', 
                             max_depth=20, 
                             min_samples_split=10, 
                             min_samples_leaf=1, 
                             min_weight_fraction_leaf=0.0, 
                             max_features='auto', 
                             max_leaf_nodes=None, 
                             min_impurity_decrease=0.0, 
                             min_impurity_split=None, 
                             bootstrap=True, 
                             oob_score=False,
                             n_jobs=4,
                             random_state=35, 
                             verbose=0, 
                             warm_start=False)
# randomly more than half of features to use
rfe = RFE(model, 20)
fit = rfe.fit(X, y)
print("Num Features: %d" % fit.n_features_)
print("Selected Features: %s" % fit.support_)
print("Feature Ranking: %s" % fit.ranking_)
```

```python
feat_list = []
for i in range(0, len(fit.support_)):
    if fit.support_[i]:
        # storing features to use in model
        feat_list.append(X.columns[i])
        print("feature to use: {0}".format(X.columns[i]))
```

```python
# using only the subset of features from previous operations (RFE)
X = df_train_cat[feat_list]
y = df_train_cat['classLabel']
```

```python
#Scikit-Learn for Modeling
from sklearn import model_selection

# using a split of 80/20
X_train, X_train_test, y_train, y_train_test = model_selection.train_test_split(X, y, test_size=0.20, random_state=15)
print ("Training and testing split was successful.")
print('Training Features Shape:', X_train.shape)
print('Training Labels Shape:', y_train.shape)
print('Testing Features Shape:', X_train_test.shape)
print('Testing Labels Shape:', y_train_test.shape)
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestClassifier
# Instantiate model with 1000 decision trees
rf_clf = RandomForestClassifier(n_estimators = 1000, max_depth=2, random_state = 42, n_jobs=4, verbose=True)
# Train the model on training data
rf_clf.fit(X_train, y_train)
```

```python
# predict on the testing set
pred = rf_clf.predict(X_train_test).tolist()
```

```python
from sklearn import metrics
# testing score
score = metrics.f1_score(y_train_test, pred)
print("f1-score: {0}".format(score))
```

on testing set, the model performed well with an f1 score of .988. I will proceed for this (not further development on model side) to predict on the validation set.

```python
import numpy as np
# validation set
validation = pd.read_csv('data/validation.csv', delimiter=';')


# apply same processing as training data
def prepare_data(dataframe):
    # convert on right format
    float_variables = ['v2','v3','v5','v6', 'v7']
    int_variables = ['v10', 'v14', 'v17']
    for var in float_variables:
        dataframe[var] = dataframe[var].apply(lambda x: float(str(x).replace(',', '.')))
    dataframe[int_variables] = dataframe[int_variables].astype(int)
    # turn class into binary
    class_label_bin = {"classLabel": {"no.": 0, "yes.": 1}}
    dataframe.replace(class_label_bin, inplace=True)
    # replace nan with median (of training) for v2
    dataframe['v2'].fillna(28.67, inplace=True)
    # v1 replacement with most common value for v1
    dataframe['v1'].fillna('b', inplace=True)
    # v4 replacement with most coomon value 
    dataframe['v4'].fillna('u', inplace=True)
    # v13 and v15 replacement
    dataframe['v15'] = pd.to_numeric(dataframe['v15'].str.replace(',','.'))
    dataframe['v13'].loc[dataframe['v4'] == 'y'] = dataframe['v13'].loc[dataframe['v4'] == 'y'].fillna(160.0)
    dataframe['v13'].loc[dataframe['v4'] == 'u'] = dataframe['v13'].loc[dataframe['v4'] == 'u'].fillna(102.0)
    dataframe['v15'].loc[dataframe['v4'] == 'y'] = dataframe['v15'].loc[dataframe['v4'] == 'y'].fillna(1600000.0)
    dataframe['v15'].loc[dataframe['v4'] == 'u'] = dataframe['v15'].loc[dataframe['v4'] == 'u'].fillna(1020000.0)
    # v16 replacement with previous built model
    pred_impute = dataframe[dataframe['v16'].isnull()]
    pred_impute = pred_impute.drop('v16', axis=1)
    pred_impute['v1'] = pred_impute['v1'].astype('category').cat.codes
    pred_impute['v4'] = pred_impute['v4'].astype('category').cat.codes
    pred_impute['v8'] = pred_impute['v8'].astype('category').cat.codes
    pred_impute['v9'] = pred_impute['v9'].astype('category').cat.codes
    pred_impute['v11'] = pred_impute['v11'].astype('category').cat.codes
    pred_impute['v12'] = pred_impute['v12'].astype('category').cat.codes
    print("-------- MISSING DATA PERCENTAGES/SUM --------")
    print(pred_impute.isnull().sum()/pred_impute.classLabel.size * 100)
    #
    # prediction with missing values
    pred_v16 = clf.predict(pred_impute).tolist()
    #
    # impute values
    pred_impute['v16'] = pred_v16
    # back to the original dataframe
    dataframe['v16'].loc[pred_impute.index] = pred_impute.v16
    #
    # turning categories into one hot encoded
    df_cat = pd.get_dummies(dataframe[['v1', 'v4', 'v8', 'v9', 'v11', 'v12', 'v16']], prefix=['v1', 'v4', 'v8', 'v9', 'v11', 'v12', 'v16']).astype(int)
    #
    # handler only
    v4_l = np.zeros(shape=df_cat['v1_a'].shape)
    df_cat['v4_l'] = v4_l
    #
    # continue process
    df_val_cat = pd.concat([dataframe, df_cat], axis=1, join_axes=[dataframe.index])
    df_val_cat = df_val_cat.drop(columns=['v1', 'v4', 'v8', 'v9', 'v11', 'v12', 'v16'], axis=1)
    #
    # select only important features
    X_val = df_val_cat.drop(columns=['classLabel'], axis=1)
    y_val = df_val_cat['classLabel']
    X_val = X_val[['v13', 'v14', 'v15', 'v17', 'v1_a', 'v1_b', 'v4_l', 'v4_u', 'v4_y', 
                   'v8_f', 'v8_t', 'v9_f', 'v9_t', 'v11_f', 'v11_t', 'v12_g', 'v12_p', 'v12_s', 
                   'v16_f', 'v16_t']]
    #
    # return X and Y
    return X_val, y_val

x_val, y_val = prepare_data(validation)
```

```python
# prediction now with the already processed dataset
val_pred = rf_clf.predict(x_val).tolist()
# testing score
score = metrics.f1_score(y_val, val_pred)
print("f1-score: {0}".format(score))
```
